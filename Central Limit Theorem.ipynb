# The Central Limit Theorem

Suppose, that we have the same $(X_k)_{k\geq0}$ sequence of Bernoulli distributed random variables and the same $(Y_i)_{i\geq0}$ sequence of Poisson distributed random variables as above.

Thus, we have the same empirical means $\bar{X}_N$ and $\bar{Y}_N$ defined by
$$\bar{X}_N = \frac{X_1 + X_2 + \dots + X_N}{N} = \frac{1}{N}\sum_{k = 1}^{N} X_k$$
and
$$\bar{Y}_N = \frac{Y_1 + Y_2 + \dots + Y_N}{N} = \frac{1}{N}\sum_{i = 1}^{N} Y_i$$

The Central Limit Theorem states that for large enough $N$, $\bar{X}_N$ will be approximately normally distributed with mean $\mathbb{E}[X_k] = p$ and variance $Var(X_k) = p(1-p)$. Similarly, for large enough $N$, $\bar{Y}_N$ will be approximately normally distributed with mean $\mathbb{E}[Y_i] = \lambda$ and variance $Var(Y_i) = \lambda$.

Using our bernoulliMean(mean,N) and poissonMean(mean,N) functions from above, let us define two simulations below that calculate and store the empirical means for both Bernoulli and Poisson sequences. We'll use $N = 1000$, running this process 10,000 times.

According to the Central Limit Theorem, graphing the outputted values of $\bar{X}_N$ and $\bar{Y}_N$ on a histogram should reveal their distribution is approximately normal for large enough $N$.

```python
def bernoulliCLT(mean,N):
    bSampleArray = [bernoulliMean(mean,N) for i in range(10000)]
    return bSampleArray

def poissonCLT(mean,N):
    pSampleArray = [poissonMean(mean,N) for i in range(10000)]
    return pSampleArray

p = 0.5
lambd = 10
N = 1000
print('CLT - 10,000 Simulations with Sequence of Bernoulli RVs:')
bArray = bernoulliCLT(p,N)
plt.hist(bArray,bins='auto')
plt.xlabel('Empirical Mean')
plt.ylabel('Frequency')
plt.title('Mean of Sequence of Bernoulli RVs (p = ' + str(p) + ', N = ' + str(N) + ')')
plt.show()

print('')

print('CLT - 10,000 Simulations with Sequence of Poisson RVs:')
pArray = poissonCLT(lambd,N)
plt.hist(pArray,bins='auto')
plt.xlabel('Empirical Mean')
plt.ylabel('Frequency')
plt.title('Mean of Sequence of Bernoulli RVs (lambda = ' + str(lambd) + ', N = ' + str(N) + ')')
plt.show()
```

As illustrated in the instances above, both $\bar{X}_N$ and $\bar{Y}_N$ resemble normal distributions for large enough $N$, just as predicted by the Central Limit Theorem.
