# The Law of Large Numbers

In this Python script, we will illustrate the Law of Large Numbers and the Central Limit Theorem for both a sequence of Bernoulli random variables and a sequence of Poisson random variables.

<b>First</b>, let's assume that $(X_k)_{k\geq0}$ is an i.i.d. sequence of Bernoulli distributed random variables with finite mean and variance, such that $\mathbb{E}[X_k] = p$ and $Var(X_k) = p(1-p)$.

Let the empirical mean $\bar{X}_N$ be defined by
$$\bar{X}_N = \frac{X_1 + X_2 + \dots + X_N}{N} = \frac{1}{N}\sum_{k = 1}^{N} X_k$$

<i>Note</i>: $X_k \sim Bernoulli(p) = Binomial(1,p)$. This is important for sampling in Python.

<b>Second</b>, let's assume $(Y_i)_{i\geq0}$ is an i.i.d. sequence of Poisson distributed random variables with finite mean and variance, such that $\mathbb{E}[Y_i] = \lambda$ and $Var(Y_i) = \lambda$.

Let the empirical mean $\bar{Y}_N$ be defined by
$$\bar{Y}_N = \frac{Y_1 + Y_2 + \dots + Y_N}{N} = \frac{1}{N}\sum_{i = 1}^{N} Y_i$$

Begin by loading the basic packages:

```python
import numpy as np
import pprint as pp
import matplotlib.pyplot as plt
```

Now, we want to define two functions that collect $N$ samples from both $(X_k)_{k\geq0}$ and $(Y_i)_{i\geq0}$ and calculate their empirical means $\bar{X}_N$ and $\bar{Y}_N$.

The bernoulliMean(mean,N) function will output $\bar{X}_N$ given a mean of $p$ and number of samples $N$.

The poissonMean(mean,N) function will ouput $\bar{Y}_N$ given a mean of $\lambda$ and number of samples $N$.

```python
def bernoulliMean(mean,N):
    bSample = np.random.binomial(1,mean,N)
    bMean = sum(bSample)/N
    return bMean

def poissonMean(mean,N):
    pSample = np.random.poisson(mean,N)
    pMean = sum(pSample)/N
    return pMean
```

By the Law of Large Numbers, we should see $\bar{X}_N \rightarrow p$ and $\bar{Y}_N \rightarrow \lambda$ as $N \rightarrow \infty$.

To test this convergence, we will implement and graph $\bar{X}_N$ and $\bar{Y}_N$ for an increasing number of $N$. Below we will use $N$ for $N \in [1,2,3,\dots,2000]$.

```python
p = 0.5
lambd = 10
#nArray = [10,50,100,500,1000,5000,10000,50000,100000]
nArray = [i for i in range(1,2001)]
bMeans = [bernoulliMean(p,N) for N in nArray]
bErrors = [np.absolute(mean-p) for mean in bMeans]
pMeans = [poissonMean(lambd,N) for N in nArray]
pErrors = [np.absolute(mean-lambd) for mean in pMeans]

print('LLN Convergence of Sequence of Bernoulli RVs:')
plt.plot(nArray,bMeans,'b')
plt.xlabel('N')
plt.ylabel('Empirical Mean')
plt.title('Means for Sequence of Bernoulli RVs (p = ' + str(p) + ')')
plt.show()
plt.plot(nArray,bErrors,'b')
plt.xlabel('N')
plt.ylabel('Absolute Error')
plt.title('Errors for Sequence of Bernoulli RVs (p = ' + str(p) + ')')
plt.show()

print('')

print('LLN Convergence of Sequence of Poisson RVs:')
plt.plot(nArray,pMeans,'b')
plt.xlabel('N')
plt.ylabel('Empirical Mean')
plt.title('Means for Sequence of Poisson RVs (lambda = ' + str(lambd) + ')')
plt.show()
plt.plot(nArray,pErrors,'b')
plt.xlabel('N')
plt.ylabel('Absolute Error')
plt.title('Errors for Sequence of Poisson RVs (lambda = ' + str(lambd) + ')')
plt.show()
```

As seen above, both $\bar{X}_N$ and $\bar{Y}_N$ converge to their means for increasing $N$, demonstrating the Law of Large Numbers for a sequence of Bernoulli RVs and a sequence of Poisson RVs.
